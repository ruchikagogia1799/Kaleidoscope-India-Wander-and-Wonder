```python
# ====== CONFIG ======
INPUT_PATH = "combined.csv"       # change if needed; falls back to in-memory `per_unique` if not found
OUTPUT_DIR = "plots"            # folder to save images
# Columns used:
# Region, State, City, Main Tourist Attraction, Google Review Rating,
# Entrance Fee (INR), Type of Attractions, Best Time (Months) or Best Time to visit
# =====================

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re

# ---------- load (CSV or in-memory per_unique) ----------
try:
    df = pd.read_csv(Path(INPUT_PATH))
    print(f"Loaded CSV: {INPUT_PATH}")
except Exception as e:
    if "per_unique" in globals():
        df = per_unique.copy()
        print("Using in-memory DataFrame: per_unique")
    else:
        raise FileNotFoundError("CSV not found and variable `per_unique` is not defined.") from e

# ---------- cleaning ----------
def _num(s, default=np.nan):
    return pd.to_numeric(s, errors="coerce").fillna(default)

df["Region"] = df["Region"].astype(str).str.strip().str.title()
df["State"]  = df["State"].astype(str).str.strip()
df["City"]   = df["City"].astype(str).str.strip()
df["Main Tourist Attraction"] = df["Main Tourist Attraction"].astype(str).str.strip()

df["Google Review Rating"] = _num(df["Google Review Rating"])
df["Entrance Fee (INR)"]   = _num(df["Entrance Fee (INR)"], default=0)
if "Number of Google Reviews" in df.columns:
    df["Number of Google Reviews"] = _num(df["Number of Google Reviews"], default=0)

# ---------- color helper: single-hue gradient (dark=high, light=low) ----------
def gradient_colors(values, cmap_name="Blues", low=0.25, high=0.95):
    vals = pd.Series(values, dtype="float64")
    if len(vals) == 0:
        return []
    vmin, vmax = float(vals.min()), float(vals.max())
    if np.isclose(vmin, vmax):  # avoid div/0
        vmax = vmin + 1.0
    normed = (vals - vmin) / (vmax - vmin)          # 0..1
    scaled = low + normed * (high - low)            # low..high
    cmap  = plt.get_cmap(cmap_name)
    return [cmap(x) for x in scaled]

# ---------- output folder ----------
out = Path(OUTPUT_DIR)
out.mkdir(parents=True, exist_ok=True)

# ========== 1) Top 5 attractions by rating (WIDE; Blues) ==========
sort_cols = ["Google Review Rating"]; ascending = [False]
if "Number of Google Reviews" in df.columns:
    sort_cols.append("Number of Google Reviews"); ascending.append(False)

top5 = (
    df.loc[df["Google Review Rating"].notna(),
           ["Main Tourist Attraction","City","State","Google Review Rating"] + (["Number of Google Reviews"] if "Number of Google Reviews" in df.columns else [])]
      .sort_values(sort_cols, ascending=ascending)
      .head(5)
      .copy()
)
top5["Label"] = (top5["Main Tourist Attraction"] + " â€” " +
                 top5["City"].fillna("") + ", " + top5["State"].fillna("")).str.strip(", ")

# reverse so best is at top
top5_plot = top5.iloc[::-1].reset_index(drop=True)
colors1 = gradient_colors(top5_plot["Google Review Rating"], cmap_name="Blues")

plt.figure(figsize=(14, 5))  # WIDE
plt.barh(top5_plot["Label"], top5_plot["Google Review Rating"], color=colors1)
plt.xlabel("Google Review Rating"); plt.ylabel("Attraction")
plt.title("Top 5 Attractions by Rating")
plt.tight_layout(); plt.savefig(out / "01_top5_attractions_by_rating.png", dpi=150, bbox_inches="tight"); plt.show()

# ========== 2) Top 10 states by number of attractions (Magma â€“ not green) ==========
state_counts = df["State"].value_counts().head(10)
colors2 = gradient_colors(state_counts.values, cmap_name="magma")  # switched from Greens -> Magma
plt.figure()
plt.bar(state_counts.index, state_counts.values, color=colors2)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Number of attractions"); plt.title("Top 10 States by Tourist Attractions")
plt.tight_layout(); plt.savefig(out / "02_top10_states_by_attractions.png", dpi=150, bbox_inches="tight"); plt.show()

# ========== 3) Attractions by Region (Oranges) ==========
region_counts = df["Region"].value_counts()
colors3 = gradient_colors(region_counts.values, cmap_name="Oranges")
plt.figure()
plt.bar(region_counts.index, region_counts.values, color=colors3)
plt.ylabel("Number of attractions"); plt.title("Attractions by Region")
plt.tight_layout(); plt.savefig(out / "03_attractions_by_region.png", dpi=150, bbox_inches="tight"); plt.show()

# ========== 4) Entrance Fee by Rating (scatter; VERY DARK Purples) ==========
sc = df.loc[(df["Google Review Rating"].notna()) & (df["Entrance Fee (INR)"] > 0),
            ["Google Review Rating","Entrance Fee (INR)"]].copy()
colors4 = gradient_colors(sc["Entrance Fee (INR)"], cmap_name="Purples", low=0.85, high=0.99)
plt.figure()
plt.scatter(sc["Google Review Rating"], sc["Entrance Fee (INR)"], c=colors4, s=40, edgecolor="none", alpha=0.85)
plt.yscale("log")
plt.xlabel("Google Review Rating"); plt.ylabel("Entrance Fee (INR, log scale)")
plt.title("Entrance Fee by Rating")
plt.tight_layout(); plt.savefig(out / "04_entrance_fee_by_rating_scatter.png", dpi=150, bbox_inches="tight"); plt.show()

# ========== 5) Count of Type of Attractions (Reds) ==========
type_counts = df["Type of Attractions"].value_counts()
colors5 = gradient_colors(type_counts.values, cmap_name="Reds")
plt.figure()
plt.bar(type_counts.index, type_counts.values, color=colors5)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count"); plt.title("Count of Type of Attractions")
plt.tight_layout(); plt.savefig(out / "05_type_of_attractions_count.png", dpi=150, bbox_inches="tight"); plt.show()

# ========== 6) Best Time to Visit  (distinct custom colors) ==========
MONTHS = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
ALIASES = {
    "jan":"Jan","january":"Jan","feb":"Feb","february":"Feb","mar":"Mar","march":"Mar",
    "apr":"Apr","april":"Apr","may":"May","jun":"Jun","june":"Jun","jul":"Jul","july":"Jul",
    "aug":"Aug","august":"Aug","sep":"Sep","sept":"Sep","september":"Sep","oct":"Oct","october":"Oct",
    "nov":"Nov","november":"Nov","dec":"Dec","december":"Dec"
}
def _m_idx(m): return MONTHS.index(m)+1
def _expand(a,b):
    ai, bi = _m_idx(a), _m_idx(b)
    return list(range(ai, bi+1)) if ai <= bi else list(range(ai,13)) + list(range(1,bi+1))

def parse_best_months(s):
    if not isinstance(s, str) or not s.strip(): return []
    t = s.lower().strip()
    if "all year" in t or "year round" in t or "year-round" in t: return list(range(1,13))
    t = t.replace("\u2013","-").replace("\u2014","-").replace(" to ","-")
    out = []
    for part in re.split(r"[,\;/]+", t):
        part = part.strip()
        if not part: continue
        if "-" in part:
            a,b = [ALIASES.get(x.strip(), "") for x in part.split("-",1)]
            if a in MONTHS and b in MONTHS: out += _expand(a,b)
        else:
            a = ALIASES.get(part, "")
            if a in MONTHS: out.append(_m_idx(a))
    return sorted(set(out))

# pick source column (Months preferred; fallback to text)
best_col = "Best Time (Months)" if "Best Time (Months)" in df.columns and df["Best Time (Months)"].notna().any() else \
           ("Best Time to visit" if "Best Time to visit" in df.columns else None)

# map months -> seasons (DJF, MAM, JJA, SON)
def month_to_season(m):
    if m in (12,1,2):  return "Winter (Decâ€“Feb)"
    if m in (3,4,5):   return "Spring (Marâ€“May)"
    if m in (6,7,8):   return "Summer (Junâ€“Aug)"
    return "Autumn (Sepâ€“Nov)"  # 9,10,11

season_counts = pd.Series(0, index=["Winter (Decâ€“Feb)","Spring (Marâ€“May)","Summer (Junâ€“Aug)","Autumn (Sepâ€“Nov)"], dtype=int)
if best_col:
    for s in df[best_col]:
        for m in parse_best_months(s):
            season_counts[month_to_season(m)] += 1

# donut with custom palette(distinct from other charts)
donut_colors = ["#2A9D8F", "#E9C46A", "#E76F51", "#264653"]  # teal, sand, coral, deep slate

plt.figure()
_ = plt.pie(
    season_counts.values,
    labels=season_counts.index,
    autopct="%1.0f%%",
    startangle=90,
    wedgeprops={"width": 0.4},
    colors=donut_colors
)
plt.axis("equal")
plt.title("Best Time to Visit â€” by Season")
plt.tight_layout(); plt.savefig(out / "06_best_time_season_donut.png", dpi=150, bbox_inches="tight"); plt.show()

print(f"Saved charts in: {out.resolve()}")
```
